```python
def summarize_dataframe(df):
    summary = {}
    
    for col in df.columns:
        # Initialize column summary
        col_summary = {}
        
        # Mean
        col_summary['mean'] = df[col].mean() if pd.api.types.is_numeric_dtype(df[col]) else None
        
        # Data type
        col_summary['dtype'] = df[col].dtype
        
        # Percentiles
        if pd.api.types.is_numeric_dtype(df[col]):
            percentiles = df[col].quantile([0.25, 0.5, 0.75]).to_dict()
            col_summary['25th_percentile'] = percentiles.get(0.25, None)
            col_summary['50th_percentile'] = percentiles.get(0.5, None)
            col_summary['75th_percentile'] = percentiles.get(0.75, None)
        else:
            col_summary['25th_percentile'] = None
            col_summary['50th_percentile'] = None
            col_summary['75th_percentile'] = None
        
        # NaN count
        col_summary['num_nans'] = df[col].isna().sum()
        
        # Add to summary
        summary[col] = col_summary
    
    # Convert summary to DataFrame
    summary_df = pd.DataFrame(summary).T
    return summary_df
df = pd.read_csv('data/clean/dataset.csv', index_col='idx')
summary = summarize_dataframe(df)
summary
```
